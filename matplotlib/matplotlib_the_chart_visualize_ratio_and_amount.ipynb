```python
# import 
import pandas as pd
import pandas.io.sql as psql
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, timedelta
```

```python
# convert timestamp to datatime 
pd.to_datetime(time.ctime(int(1566302685420/1000000))))
```

```python
# find the all days between two time values
start_date = '2019-07-28'
end_date = '2019-08-02'
ls_start_date = [int(i) for i in start_date.split('-')]
ls_end_date = [int(i) for i in end_date.split('-')]

sdate = date(ls_start_date[0], ls_start_date[1], ls_start_date[2])
edate = date(ls_end_date[0], ls_end_date[1], ls_end_date[2])

delta = edate - sdate       # as timedelta
ls_date = []
for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    ls_date.append(day)
ls_date
```

```python
# data
date = ls_date
version = ['1_1', '1_2', '1_3', '1_4', '1_5', '1_6']
ty_le_1 = [15.62, 12.55, 14.81, 14.62, 13.78, 11.68]
so_luong_1 = [343, 260, 282, 354, 337, 291]
ty_le_2 = [62.68, 60.38, 52.48, 60.73, 62.61, 34.71]
so_luong_2 = [215, 157, 148, 215, 211, 101]
ty_le_3 = [37.32, 39.62, 47.52, 39.27, 37.39, 65.29]
so_luong_3 = [128, 103, 134, 139, 126, 190]
so_luong_4 = [ 950,  855,  787, 1125, 1048, 1058]
```

```python
# ratio chart pie follow the day
sizes = [ty_le_2[0], ty_le_3[0]]
ls_ratio = []
for i in range(len(sizes)):
    ls_ratio.append(str(sizes[i]) + '%')
    
plt.figure(figsize= (5,5))
plt.title('ratio chart')
plt.pie(sizes, labels=ls_ratio, colors=['blue', 'yellow'], startangle=90, autopct='%.1f%%')
plt.xlabel(date[0])
plt.show()
```

```python
ls_ratio = ['ty_le_1', 'ty_le_2', 'ty_le_3']
ls_count = ['so_luong_1', 'so_luong_2', 'so_luong_3']
labels = np.arange(len(ls_ratio))
size_ratio = [ty_le_1[0], ty_le_2[0], ty_le_3[0]]
size_count = [so_luong_1[0], so_luong_2[0], so_luong_3[0]]

label_col_count = [str(x) for x in size_count]
label_col_ratio = [str(x)+'%' for x in size_ratio]

plt.figure(figsize= (13,4))
plt.subplot(1, 2, 1)
plt.bar(labels, size_ratio, width= 0.3)
plt.xticks(labels, ls_ratio, rotation=45, fontsize = 13)
plt.title('ratio in date: {}'.format(date[0]), fontsize = 15)
# Text on the top of each barplot
for i in range(len(labels)):
    plt.text(x = labels[i] , y = size_ratio[i], s = label_col_ratio[i], size = 13)
    
plt.subplot(1, 2, 2)
plt.bar(labels, size_count, width= 0.3)
plt.xticks(labels, ls_count, rotation=45, fontsize = 13)
plt.title('amount in date: {}'.format(date[0]), fontsize= 15)
# Text on the top of each barplot
for i in range(len(labels)):
    plt.text(x = labels[i] , y = size_count[i], s = label_col_count[i], size = 13)

    
plt.show()
```
